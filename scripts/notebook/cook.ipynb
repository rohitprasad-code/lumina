Method 2: Official Cloud Control (tuya-iot-py-sdk)
> This is the official and most reliable cloud method. It requires an **Access ID** and **Access Secret** from the [Tuya IoT Platform](https://iot.tuya.com/).

```python
from tuya_iot import TuyaOpenAPI, AuthType

# Credentials from Tuya IoT Platform
ENDPOINT = "https://openapi.tuyain.com"
ACCESS_ID = os.getenv("TUYA_ACCESS_ID")
ACCESS_SECRET = os.getenv("TUYA_ACCESS_SECRET")
USERNAME = os.getenv("TUYA_USERNAME")
PASSWORD = os.getenv("TUYA_PASSWORD")
```

```python
# Initialize API
openapi = TuyaOpenAPI(ENDPOINT, ACCESS_ID, ACCESS_SECRET, AuthType.SMART_HOME)
openapi.connect(USERNAME, PASSWORD, "91", "smart_life")
```

```python
# Get Device List
response = openapi.get("/v1.0/iot-01/associated-users/devices")
if response.get("success"):
    devices = response.get("result", {}).get("list", [])
    print(f"Official SDK Success! Found {len(devices)} devices.")
    for dev in devices:
        print(f"- {dev['name']} (ID: {dev['id']})")
else:
    print("Official SDK Error:", response.get("msg"))
```
